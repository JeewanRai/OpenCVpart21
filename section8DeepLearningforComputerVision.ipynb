```python
import numpy as np
from numpy import genfromtxt # generates an array from text file
```

```python
# Specify the file path
data_path = r'P:\\Pi OpenCV  programming\\banknotedata.txt'

# Function to print a specific line number for debugging
def print_line(file_path, line_number):
    with open(file_path, 'r') as file:
        for current_line_number, line in enumerate(file, start=1):
            if current_line_number == line_number:
                print(f"Line {line_number}: {line.strip()}")
                break

# Print the problematic line
line_of_interest = 1373
print_line(data_path, line_of_interest)

# Attempt to read the file, skipping bad lines
try:
    data = np.genfromtxt(data_path, delimiter=',', invalid_raise=False)
    print("Data loaded successfully.")
    print(data[:5])
except Exception as e:
    print("An error occurred while loading data:", str(e))
```

```python
data
```

```python
# saperating labels 
labels = data[:, 4]
labels
```

```python
features = data[:, 0:4]
features
```

```python
# saperating data for traing and testing, X and y are methamatical notation used in ML
X = features
y = labels

# spliting data in training and test data set, split X and y data into training and test set randomly
from sklearn.model_selection import train_test_split

# split features and labels into training and test set into 33% and 42 is standard value
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train
```

```python
# to make neural network perform better
from sklearn.preprocessing import MinMaxScaler

# create scalar object
scaler_object = MinMaxScaler()

# fit the scalar on training data
scaler_object.fit(X_train)


```

```python
scaled_X_train = scaler_object.transform(X_train)

scaled_X_test = scaler_object.transform(X_test)

scaled_X_train.max()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


model = Sequential()

model.add(Dense(4, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(scaled_X_train, y_train, epochs=50,verbose = 2, batch_size=10)
```

```python


model.metrices_names
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Assuming your model has been defined, compiled, and trained already:
# model = ...

# Use the predict method to get the output probabilities
predictions = model.predict(scaled_X_test)

```

```python

# Convert probabilities to class labels (0 or 1)
predicted_classes = (predictions > 0.5).astype(int)

# Now, you can use predicted_classes to evaluate the model
from sklearn.metrics import classification_report, confusion_matrix

# Assuming y_test contains the true class labels
print(confusion_matrix(y_test, predicted_classes))
```

```python
print(classification_report(y_test, predicted_classes))
```

## Keras Convolution Neural Netowrk with MNIST

```python
from keras.datasets import mnist
```

```python
# load MNIST data using tuple unpacking
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
x_train.shape
```

```python
# grab the single image from the first index
single_image = x_train[0]
single_image
```

```python
plt.imshow(single_image, cmap='gray_r')
```

```python
# feeding the data in this format/category data in this way the network will not understand
y_train
```

```python
y_train.shape
```

```python
from keras.utils import to_categorical

# convert the labels into categorical data, 10 indicates number of category
y_train_categorical = to_categorical(y_train, 10)
y_test_categorical = to_categorical(y_test, 10)
```

```python
y_test_categorical
```

```python
y_train_categorical[0]
```

```python
# processing x data, its not normalize, should normalize
single_image.max()
```

```python
x_train = x_train / x_train.max() # same as dividing by /255
```

```python
x_test = x_test / x_test.max() # same as dividing by /255
```

```python
scaled_image = x_train[0]
scaled_image
```

```python
scaled_image.max()
```

```python
plt.imshow(scaled_image, cmap='gray_r')
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(60000, 28, 28, 1)
x_train.shape
```

```python
x_test = x_test.reshape(10000, 28, 28, 1)
x_test.shape
```

```python
from keras.models import Sequential

from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D
```

```python
# create model, creating sequencial onject
model = Sequential()

# CONVOLUTIONAL LAYERS
# for this images we can use 32 as difault filter otherwise for complex image we need to use higher filter size
model.add(Conv2D(filters = 32, kernel_size = (4,4),input_shape = (28, 28, 1), activation='relu'))

# POOLING LAYERS
model.add(MaxPooling2D(pool_size=(2,2)))

# converting Convolution and MAXPooling layers such that single dense layer can understand
model.add(Flatten()) # 2D------> 1D

# DENSE LAYERS
model.add((Dense(128, activation='relu'))) # fully connected layer, experiment with 128 values

model.add(Dense(10, activation='softmax')) # output layer, 10 is difault value for 10 digits

model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.summary()
```

```python
model.fit(x_train, y_train_categorical, epochs=10)
```

```python
# grab and evaluate the model
model.metrics_names
```

```python
model.evaluate(x_test, y_test_categorical)
```

```python
#predict class using images not seen by model before
from sklearn.metrics import classification_report

predictions = model.predict(x_test)
```

```python
y_test_categorical
```

```python
predictions
```

```python
y_test
```

```python
import numpy as np

# Convert the probabilities to class labels
predicted_classes = np.argmax(predictions, axis=1)

# Now you can use 'predicted_classes' with 'y_test' in your classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted_classes))

```

## Keras Convolution Neural Netowrk with CIFAR-10, COLOR IMAGES

```python
from keras.datasets import cifar10
```

```python
# loead data, using tuple upacking
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

```python
# pre processing
x_train.max()
```

```python
x_train = x_train / x_train.max() # same as dividing by /255
x_test = x_test / x_test.max() # same as dividing by /255
```

```python
x_test.shape
```

```python
y_train
```

```python
from keras.utils import to_categorical
y_train_categorical = to_categorical(y_train, 10)
y_test_categorical = to_categorical(y_test, 10)
```

```python
# Build our model
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential()

# CONVOLUTIONAL LAYERS
model.add(Conv2D(filters = 32, kernel_size = (4,4), input_shape = (32, 32, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2,2)))

# CONVOLUTION LAYERS, the image is complex so another convolution layer is good idea
model.add(Conv2D(filters = 32, kernel_size = (4,4), input_shape = (32, 32, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten()) # 2D------> 1D

# DENSE LAYERS
model.add(Dense(256, activation='relu')) # fully connected layer, experiment with 128 values

model.add(Dense(10, activation='softmax')) # output layer, 10 is difault value for 10 digits

model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train_categorical, verbose = 1, epochs=10)
```

```python
model.metrics_names
```

```python
model.evaluate(x_test, y_test_categorical)
```
